```python
# STEP 2: Preprocess text manually without using external libraries

def preprocess(text):
    # Step 1: lowercase
    text = text.lower()

    # Step 2: remove punctuation (manual)
    punctuation = ".,!?;:-()"
    for p in punctuation:
        text = text.replace(p, "")

    # Step 3: split into words
    words = text.split()

    return words

# TEST the preprocess function
for r in reviews:
    print(preprocess(r))
```

```python
# STEP 3: Build a simple sentiment model manually (no sklearn)

# Dictionaries to store word frequencies
pos_words = {}
neg_words = {}

# Go through each review and update dictionary counts
for review, label in zip(reviews, labels):
    words = preprocess(review)

    if label == "pos":
        for w in words:
            pos_words[w] = pos_words.get(w, 0) + 1

    elif label == "neg":
        for w in words:
            neg_words[w] = neg_words.get(w, 0) + 1

# Print a few examples
print("Positive word frequencies:", pos_words)
print("\nNegative word frequencies:", neg_words)
```

```python
def predict_sentiment(text):
    words = preprocess(text)

    pos_score = 0
    neg_score = 0

    # Count only presence, not frequency
    for w in words:
        if w in pos_words:
            pos_score += 1
        if w in neg_words:
            neg_score += 1

    if pos_score > neg_score:
        return "Positive Polarity"
    elif neg_score > pos_score:
        return "Negative Polarity"
    else:
        return "Neutral"
```

```python

# STEP 5: Real-time user input prediction

user_text = input("Enter your sentence: ")
result = predict_sentiment(user_text)
print("Sentiment:", result)

```
